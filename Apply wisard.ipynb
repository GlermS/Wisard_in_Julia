```julia
using CSV, DataFrames, Tables, StatsBase
include("WisardPkg.jl")
```

```julia
function accuracy(guesses::Vector, truth::Vector)
    hits = sum([y1==y2 ? 1 : 0 for (y1, y2) in zip(guesses, truth)])
    return hits/length(truth)
end
```

```julia
dfi = DataFrame(CSV.File("./iris_dataset.csv"))
dfi = dfi[sample(1:size(dfi)[1], 150), :]

# Dataset de treino
train_idx = 1:100
test_idx = []
df = dfi[train_idx, :]

# Um termometro por metrica
thermos = [WisardPkg.Thermometer(10, minimum(df[:,i]), maximum(df[:,i])) for i in 1:4]
```

```julia
inputs = Tables.rowtable(df[:, 1:4])
output = [x for x in df[:,5]]

X = [[WisardPkg.encode(thermos[i], x[i]) for i in 1:4] for x in inputs]
y = output

model = WisardPkg.Wisard(40, 9, unique(output), true)
WisardPkg.fit(model, X, y)

prediction = WisardPkg.predict(model, X)

# Teste no dataset completo
output2 = [x for x in dfi[:,5]]
inputs2 = Tables.rowtable(dfi[:, 1:4])

X = [[WisardPkg.encode(thermos[i], x[i]) for i in 1:4] for x in inputs2]
y = output2

prediction2 = WisardPkg.predict(model, X)



accuracy(prediction, output), accuracy( prediction2, output2)
```

```julia
# Teste no subconjunto no visto no treinamento

output2 = [x for x in dfi[100:150 ,5]]
inputs2 = Tables.rowtable(dfi[100:150 , 1:4])

X = [[WisardPkg.encode(thermos[i], x[i]) for i in 1:4] for x in inputs2]
y = output2

prediction2 = WisardPkg.predict(model, X)

accuracy( prediction2, output2)
```
